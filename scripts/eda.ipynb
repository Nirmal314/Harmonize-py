```python
categorical_cols = ['tempo_cat', 'loudness_cat', 'valence_cat']
df[categorical_cols].describe()
```

```python
numerical_cols = ['valence','energy','danceability','instrumentalness','liveness'
                  ,'speechiness','acousticness','loudness_scaled','tempo_scaled']

def plot_histograms(df, columns, rows, cols):
    fig, axes = plt.subplots(rows, cols, figsize=(20, 4*rows))
    axes = axes.flatten()
    
    for i, col in enumerate(columns):
        sns.histplot(data=df, x=col, ax=axes[i], kde=True)
        axes[i].set_title(f'Distribution of {col}')
    
    plt.tight_layout()

    plt.show()

plot_histograms(df, numerical_cols, 3, 3)
```

```python
def plot_correlation_heatmap(df, columns):
    plt.figure(figsize=(12, 10))
    sns.heatmap(df[columns].corr(), annot=True, cmap='coolwarm', vmin=-1, vmax=1)
    plt.title('Correlation Heatmap of Audio Features')
    plt.show()

plot_correlation_heatmap(df, numerical_cols)
```

```python

def plot_countplots(df, categorical_cols, rows, cols):
    fig, axes = plt.subplots(rows, cols, figsize=(15, 5*rows))
    axes = np.ravel(axes) 
    
    for i, cat_col in enumerate(categorical_cols):
        sns.countplot(data=df, x=cat_col, ax=axes[i])
        axes[i].set_title(f'Distribution of {cat_col}')
        axes[i].tick_params(axis='x', rotation=45)  

    for j in range(len(categorical_cols), len(axes)):
        axes[j].axis('off')
    
    plt.tight_layout()
    plt.show()


plot_countplots(df, categorical_cols, 1, 3)
```

```python
def plot_boxplots(df, categorical_cols, rows, cols, y_col='danceability'):
    fig, axes = plt.subplots(rows, cols, figsize=(15, 5*rows))
    axes = np.ravel(axes)
    
    for i, cat_col in enumerate(categorical_cols):
        sns.boxplot(data=df, x=cat_col, y=y_col, ax=axes[i])
        axes[i].set_title(f'Danceability by {cat_col}')
        axes[i].tick_params(axis='x', rotation=45)

    for j in range(len(categorical_cols), len(axes)):
        axes[j].axis('off')
    
    plt.tight_layout()
    plt.show()

plot_boxplots(df, categorical_cols, 1, 3)
```
